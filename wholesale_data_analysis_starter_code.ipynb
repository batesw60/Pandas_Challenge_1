## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df["total_unit_cost"] = df.unit_price * df.qty
df.head()
```

```python
df["total_weight"] = df.qty * df.unit_weight
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df["shipping_price"] = df.total_weight.apply(lambda x: 7 * x if x > 50 else 10 * x)  
df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df["total_price"] = round((df.total_unit_cost + df.shipping_price) * 1.0925, 2)
df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df["total_cost"] = df.unit_cost * df.qty + df.shipping_price
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
df["profit"] = df.total_price - df.total_cost
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
def client_total_spend(client_id):
    temp_df = df[df.client_id == client_id]
    return temp_df["total_price"].sum()
```

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
for client in top_clients_list:
    print(f"Client Number: {client}: ${client_total_spend(client):,.2f}")
```

```python
# Create a summary df with all the clients
summary_df = df[["client_id", "qty", "shipping_price", "total_price", "total_cost", "profit"]]
summary_df.head()
```

```python
summed_summary_df = summary_df.groupby("client_id", as_index=False).sum()
summed_summary_df
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_five_summary_df = summed_summary_df[summed_summary_df.client_id.isin(top_clients_list)]
top_five_summary_df.reset_index(drop=True, inplace=True)
top_five_summary_df
```

```python
top_five_summary_df.sort_values(by= ["profit"], inplace=True)
top_five_summary_df
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
money_columns_list = ["shipping_price", "total_price", "total_cost", "profit"]
```

```python
# Define a function that converts a dollar amount to millions.
def convert_to_millions(dollar_amount):
    return f"${(dollar_amount / 1000000):,.2f}M"

```

```python
# Apply the currency_format_millions function to only the money columns. 
for col in money_columns_list:
    top_five_summary_df[col] = top_five_summary_df[col].apply(convert_to_millions)

top_five_summary_df
```

```python
# Rename the columns to reflect the change in the money format. 
final_clean_df = top_five_summary_df.rename(columns={
    'client_id': 'Client ID',
    'qty': 'Units',
    'shipping_price': 'Shipping (millions)',
    'total_price': 'Total Revenue (millions)',
    "total_cost": "Total Cost (millions)",
    "profit": "Total Profit (millions)"
})

final_clean_df

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
final_clean_df.sort_values("Total Profit (millions)", ascending=False, inplace=True)
final_clean_df
```
