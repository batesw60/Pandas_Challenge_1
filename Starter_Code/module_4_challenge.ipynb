## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df["total_cost"] = df.unit_price * df.qty
df.head()
```

```python
# Create a column for total weight
df["total_weight"] = df.unit_weight * df.qty
df.head()
```

```python
# function for shipping price
def get_shipping_price(weight):
    if weight > 50:
        return weight * 7
    else:
        return weight * 10
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
df["shipping_price_($)"] = df.total_weight.apply(get_shipping_price)
df.head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
sales_tax = 1.0925

df["line_price"] = (df.total_cost + df["shipping_price_($)"]) * sales_tax
df.line_price = df.line_price.round(2)
df.head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df["line_cost"] = (df.unit_cost * df.qty) + df["shipping_price_($)"]
df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
df["line_profit"] = df.line_price - df.line_cost
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
df.client_id.value_counts().head()
```

```python
def client_total_spend(client_id):
    temp_df = df[df.client_id == client_id]
    return temp_df["line_price"].sum()
```

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
print(f"33615: ${client_total_spend(33615):,.2f}")
print(f"66037: ${client_total_spend(66037):,.2f}")
print(f"46820: ${client_total_spend(46820):,.2f}")
print(f"38378: ${client_total_spend(38378):,.2f}")
print(f"24741: ${client_total_spend(24741):,.2f}")
```

```python
summary_df = df[["client_id", "qty", "shipping_price_($)", "line_price", "line_cost", "line_profit"]]
summary_df.head()
```

```python
grouped_client_summary = summary_df.groupby(["client_id"]).sum()
grouped_client_summary.head()
```

```python
grouped_client_summary = grouped_client_summary.reset_index()
grouped_client_summary.head()
```

```python
final_df = grouped_client_summary.loc[grouped_client_summary.client_id.isin([24741, 38378, 66037, 46820, 33615])]
final_df
```

```python

final_df = final_df.sort_values(by=["line_profit"], ascending=False)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
final_df = final_df.reset_index(drop="True")
try_df = final_df
final_df
```

```python
try_df[["shipping_price_($)", "line_price", "line_cost", "line_profit"]] = final_df[["shipping_price_($)", "line_price", "line_cost", "line_profit"]] / 1000000
try_df
```

```python
try_df[["shipping_price_($)", "line_price", "line_cost", "line_profit"]] = try_df[["shipping_price_($)", "line_price", "line_cost", "line_profit"]].map("${:,.2f}M".format)
```
